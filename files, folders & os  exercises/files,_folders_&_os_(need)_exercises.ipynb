```python
month_dict={
'jan': '01', 'feb': '02', 'mar': '03',
'apr': '04', 'may': '05', 'jun': '06',
'jul': '07', 'aug': '08', 'sep': '09',
'oct': '10', 'nov': '11', 'dec': '12'
}
```

```python
sample = "oct-08_singapore.txt"
print(sample.split("-"))
```

```python
sample = "oct-08_singapore.txt"
month = sample.split("-")[0]                   #split to 'oct' and '08_singapore.txt'
month = month_dict[month]                      #matching months to dictionary 'oct' to '10'
combined = month + "-" + sample.split("-")[1]  #merge '10' + '-' + '08_singapore.txt'
final = combined.replace("_","-")              #replace '_' with '-'
print(final)
```

```python
import glob
```

```python
import os
```

```python
old_text_files = glob.glob('os-collaboration-exercise-data/*.txt')

def rename_my_file(old_file_name):
    month = old_file_name.split("-")[0]
    month = month_dict[month]
    combined = month + "-" + old_file_name.split("-")[1]
    new_file_name = combined.replace("_","-")
    return new_file_name

for old_text_file in old_text_files:
    directory = old_text_file.split("\\")[0]
    file_name = old_text_file.split("\\")[1]
    new_text_file = rename_my_file(file_name)
    new_text_file = directory + "\\" + new_text_file
    os.rename(old_text_file, new_text_file)
```

### Task 4 Updated

```python
import os
import glob

old_text_files = glob.glob('os-collaboration-exercise-data/*.txt')

def rename_my_file(old_file_name):
    month = old_file_name.split("-")[0]
    month = month_dict[month]
    combined = month + "-" + old_file_name.split("-")[1]
    new_file_name = combined.replace("_","-")
    return new_file_name

for old_text_file in old_text_files:
    directory, file_name = os.path.split(old_text_file)
    new_text_file = os.path.join(directory, rename_my_file(file_name))
    os.rename(old_text_file, new_text_file)
```
